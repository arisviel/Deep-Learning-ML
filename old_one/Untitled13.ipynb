```python
with open('py.txt') as inp:
    inlet=inp.read().strip().lower().split()
x=0
y=''
for i in inlet:
    if inlet.count(i)>x:
        x=inlet.count(i)
        y=i
    if inlet.count(i)==x:
        if i>y:
            x=inlet.count(i)
            y=i
print(y, x)
```
